```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv('vgsales.csv')

df.head()
```

```python
data_types = df.dtypes
print(data_types)
```

```python
df['Year_of_Release'] = df['Year_of_Release'].fillna(0).astype(int)
```

```python
df['User_Score'] = pd.to_numeric(df['User_Score'], errors='coerce')
```

```python
df['Critic_Count'] = df['Critic_Count'].fillna(0).astype(int)
df['User_Count'] = df['User_Count'].fillna(0).astype(int)
```

```python
print("\nMissing values in the dataset:")
print(df.isnull().sum())
```

```python
print("\nStatistical details of the dataset:")
print(df.describe())
```

```python
# Remove rows where 'Name' or 'Genre' is missing
df.dropna(subset=['Name', 'Genre'], inplace=True)

# Replace missing 'Publisher' values with 'Unknown'
df['Publisher'].fillna('Unknown', inplace=True)

# Display the DataFrame to confirm changes
df.head()
```

```python
platform_counts = df['Platform'].value_counts()
print("\nNumber of games per platform:")
print(platform_counts)
```

```python
genre_counts = df['Genre'].value_counts()
print("\nNumber of games per genre:")
print(genre_counts)
```

```python
plt.figure(figsize=(10, 6))
sns.distplot(df['Global_Sales'].dropna(), kde=False, bins=50)
plt.title('Distribution of Global Sales')
plt.show()
```

```python
# Correlation matrix heatmap
plt.figure(figsize=(10, 8))
corr = df.corr()
sns.heatmap(corr, annot=True, fmt=".2f")
plt.title('Correlation Matrix')
plt.show()

# Scatter plot of global sales vs. critic scores
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Critic_Score', y='Global_Sales', data=df)
plt.title('Global Sales vs. Critic Score')
plt.show()
```
